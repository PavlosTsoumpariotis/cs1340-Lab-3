```python
import numpy

print(numpy.__version__)

import datetime


class Stock:

    def __init__(self, filename):
        self.filename = filename
        self.data_list = self.get_data(self.filename)

    def get_data(self, filename):

        with open(filename, "r") as f:
            next(f)
            data = []
            for row in f:
                row = row.strip("\n")
                row = row.split(",")
                data.append(row)
            return data

    def calculate_moving_average(self, windowsize):

        movingavg = [] #a list that stores moving averages
        temporarylist = [] #a list that stores the closing prices
        counter = 0
        for rows in self.data_list:
            if counter < windowsize:
                counter += 1
                a = float(rows[5])
                temporarylist.append(a)
                movingaverage = sum(temporarylist) / counter
                movingavg.append(movingaverage)
            else:
                a = float(rows[5])
                temporarylist.append(a)
                movingaverage = sum(temporarylist) / counter
                movingavg.append(movingaverage)
                temporarylist.pop(0) #removes first element of temporary list each time counter increases
        return movingavg

    def get_returns(self, buy_price, volume, end_date):
        
        dates = []
        for rows in self.data_list:
            for columns in range(1):
                dates.append(rows[0])
        month_day = datetime.datetime.strptime(end_date, "%Y-%m-%d").date()
        a = dates.index(str(month_day))
        adjclose = self.data_list[a][5]
        profit = (float(adjclose) - float(buy_price)) * float(volume)

        return float(profit)


filename = "C:/Lab2/GOOG.csv"
data_list = Stock(filename)
print(data_list.data_list)
print("This function returns a data list!")
print(data_list.calculate_moving_average(5))
print("This function returns a list of the moving averages!")
print(data_list.get_returns(100, 6, '2019-9-13'))
print("This functions calculates return on stock!")
```
